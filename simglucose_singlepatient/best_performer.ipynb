```python
import numpy as np
from pathlib import Path

outdir = Path("./teacher_models/postitve/adult-010")  # change this

eval_file = outdir / "eval" / "evaluations.npz"
data = np.load("/home/elisalaegs/GGSpeciale/GGSpeciale/simglucose_singlepatient/teacher_models/positive/adult-010/eval/evaluations.npz")

timesteps = data["timesteps"]
results = data["results"]  # shape: [n_eval_calls, n_eval_episodes]

mean_rewards = results.mean(axis=1)
std_rewards = results.std(axis=1)

best_idx = int(np.argmax(mean_rewards))

print("Best model was saved at:")
print(f"  timestep: {timesteps[best_idx]}")
print(f"  mean_reward: {mean_rewards[best_idx]}")
print(f"  std_reward: {std_rewards[best_idx]}")
print()
print("All evals:")
for t, m, s in zip(timesteps, mean_rewards, std_rewards):
    marker = "<-- BEST" if t == timesteps[best_idx] else ""
    print(f"{int(t):>10} | mean={m:>12.3f} | std={s:>12.3f} {marker}")
```
